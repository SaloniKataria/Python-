## Exercise 1

```python
import pandas as pd
```

```python
#Exercise 1.1
df=pd.read_csv('UIC2016Basketball.csv')
```

```python
#Exercise 1.2

names_col=['Date','Opponent','UIC Score','Opp Score','UIC Field Goal Percentage','Opp Field Goal Percenatage',
                         'UIC 3 point Field Goal Percentage','Opp 3 point Field Goal Percentage','UIC Rebound','Opp Rebound',
                         'UIC Assists','Opp Assists']


df=pd.read_csv('UIC2016Basketball.csv',names=names_col)
print(df)
```

```python
#Exercise 1.3
df = df.drop(labels=0, axis=0)
df.fillna('-')
```

```python
#Exercise 1.4
df.dtypes
```

```python
#Exercise 1.5

df['Date'] = pd.to_datetime(df['Date'])
new_df = df[(df['Date'] > "01/01/2016") & (df['Date'] < "12/31/2016")]
print('Unique teams =',new_df['Opponent'].unique())
print('Count =',new_df['Opponent'].nunique())
```

```python
#Exercise 1.6
df['UIC Score'] = pd.to_numeric(df['UIC Score'])

filter_df = df[df['UIC Score'] > 65]
print(filter_df)

# Answer : There are 15 rows with UIC Score > 65
```

```python
#Exercise 1.7

df['Opp Score'] = pd.to_numeric(df['Opp Score'])
df.loc[df['UIC Score'] == df['Opp Score'], 'match'] = 'tie'
df.loc[df['UIC Score'] > df['Opp Score'], 'match'] = 'UIC_Won'
df.loc[df['UIC Score'] < df['Opp Score'], 'match'] = 'UIC_Lost'

df.groupby(['match']).size()

#Ties: 0
#Wins: 5
#Lost: 25
```

## Exercise 2

```python
n_col=['Customer ID', 'Customer Name', 'Customer Surname', 'Gender', 'Age', 'Region', 'Job Classification', 'Date Joined', 'Balance']
Data = pd.read_csv("customer-savings.txt", sep=',',names=n_col)
nData = pd.DataFrame(Data)
print(nData)
```

```python
mBal = nData[(nData['Gender'] == 'Male')]
print(mBal)
mBal['Balance'].mean()

#Answer: The average balance for male is 40018.00891454967
```

```python
fBal = nData[(nData['Gender'] == 'Female')]
print(fBal)
fBal['Balance'].mean()
 
#Answer: The average balance for female is 39471.89511627909
```

```python
wCol = nData[(nData['Job Classification'] == 'White Collar') & (nData['Region'] == 'England')]
print(wCol)
wCol['Balance'].mean()

#Answer: The average balance for White Collar in England is 39156.30570952698
```

```python
bCol = nData[(nData['Job Classification'] == 'Blue Collar') & (nData['Region'] == 'England')]
print(bCol)
bCol['Balance'].mean()

#Answer: The average balance for Blue Collar in England is 38567.841569767465
```

```python
cstatus = pd.read_csv('customer-status.csv')
```

```python
sales = pd.read_csv('sales.csv')
```

```python
df1 = pd.DataFrame(cstatus)
print(df1)
```

```python
df2 = pd.DataFrame(sales)
print(df2)
```

```python
pd.merge(df1,df2, on = 'account number')

#Answer - No. of Rows: 118
```

```python
pd.merge(df1,df2, on = 'account number', how = 'outer')

#Answer - No. of Rows: 142
```

```python
pd.merge(df1,df2, on = 'account number', how = 'left')

#Answer - No. of Rows: 118
```

```python
pd.merge(df2,df1, on = 'account number', how = 'left')

#Answer - No. of Rows: 142
```
