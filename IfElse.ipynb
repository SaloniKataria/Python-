```python
import os
os.getcwd()
os.chdir('C:\\Users\\aksha\\OneDrive\\Desktop\\Assignment3')
inp=open('retail.txt','r')

popular_pairs = {}
popular_products = {}


for i in inp:
    i = i.replace('\n','')     
    i = i.split(' ')
    start_id = i[0]
    first_pr = i[1]
    second_prd = i[2]
    if second_prd != '0':         
        pair_ = (first_pr,second_prd)
        if pair_ not in popular_pairs.keys():
            popular_pairs[pair_]=1
        else:
            popular_pairs[pair_]+=1
        
        if first_pr not in popular_products.keys():  
            popular_products[first_pr]=1
        else:
            popular_products[first_pr]+=1
            
        if second_prd not in popular_products.keys():
            popular_products[second_prd]=1
        else:
            popular_products[second_prd]+=1
            
    else:
        if first_pr not in popular_products.keys():
            popular_products[first_pr]=1
        else:
            popular_products[first_pr]+=1
        


sorted_pop_pairs = sorted(popular_pairs.items(),key= lambda kv:kv[1],reverse=True)[:11]
sorted_pop_products = sorted(popular_products.items(),key= lambda kv:kv[1],reverse=True)[:11]




output=open("output.txt","w")
output.write("Top 10 Purchased items:\n")
a = str(sorted_pop_products)
output.write(a)

output.write("Top 10 Co-purchased items:\n")
b = str(sorted_pop_pairs)
output.write(b)

output.close()
inp.close()
```
